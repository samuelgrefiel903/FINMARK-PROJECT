```python
# ============================================================
# END-TO-END EDA + PREPROCESSING + FEATURE ENGINEERING (WINDOWS)
# Fixes common file/path + encoding + dtype + plotting issues
# ============================================================

import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------------
# 0) SETTINGS (WINDOWS PATHS)
# ----------------------------
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 160)
plt.rcParams["figure.figsize"] = (10, 5)

# ‚úÖ Your Windows file path (use raw string r"...")
DATA_PATH = r"C:\Users\lawre\Downloads\customers_data (1).csv"

# ‚úÖ Output folder (you can change this)
OUT_DIR = r"C:\Users\lawre\Downloads\eda_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

CLEAN_PATH    = os.path.join(OUT_DIR, "customers_cleaned.csv")
FEATURES_PATH = os.path.join(OUT_DIR, "customers_features.csv")
REPORT_PATH   = os.path.join(OUT_DIR, "eda_report.txt")

# Optional: if you know your target column name, set it here.
# Example: TARGET_COL = "Purchased"
TARGET_COL = None


# ----------------------------
# 1) LOAD DATA (ROBUST)
# ----------------------------
def robust_read_csv(path: str) -> pd.DataFrame:
    """
    Tries multiple encodings and separators. Fixes the #1 causes of read errors.
    """
    if not os.path.exists(path):
        raise FileNotFoundError(
            f"File not found:\n{path}\n\n"
            "Tips:\n"
            "- Check spelling and spaces in the filename\n"
            "- Make sure the file is really in Downloads\n"
            "- Try right click file > Properties to confirm name"
        )

    read_attempts = [
        {"encoding": "utf-8", "sep": ","},
        {"encoding": "utf-8-sig", "sep": ","},
        {"encoding": "cp1252", "sep": ","},
        {"encoding": "latin1", "sep": ","},
        # If file is semicolon-separated (common in some exports):
        {"encoding": "utf-8", "sep": ";"},
        {"encoding": "cp1252", "sep": ";"},
    ]

    last_err = None
    for opts in read_attempts:
        try:
            df = pd.read_csv(path, **opts)
            # If it loads but becomes 1 column, it's likely wrong separator
            if df.shape[1] == 1 and opts["sep"] == ",":
                # try next attempt
                continue
            print(f"‚úÖ Loaded CSV with encoding={opts['encoding']} sep='{opts['sep']}'")
            print("Shape:", df.shape)
            return df
        except Exception as e:
            last_err = e

    raise RuntimeError(f"Could not read the CSV. Last error:\n{last_err}")


df_raw = robust_read_csv(DATA_PATH)
print("\n--- HEAD ---")
display(df_raw.head(10))
print("\n--- INFO ---")
df_raw.info()


# ----------------------------
# 2) EDA
# ----------------------------
def basic_report(df: pd.DataFrame, report_path: str):
    lines = []
    lines.append("===== BASIC DATA REPORT =====")
    lines.append(f"Shape: {df.shape}")
    lines.append("\nColumns:")
    lines.append(", ".join([str(c) for c in df.columns]))

    lines.append("\n\nDtypes:")
    lines.append(df.dtypes.astype(str).to_string())

    miss = df.isna().sum().sort_values(ascending=False)
    miss_nonzero = miss[miss > 0]
    lines.append("\n\nMissing values (non-zero):")
    lines.append(miss_nonzero.to_string() if len(miss_nonzero) else "None")

    lines.append(f"\n\nDuplicate rows: {df.duplicated().sum()}")

    nunique = df.nunique(dropna=False).sort_values()
    lines.append("\n\nnunique (sorted):")
    lines.append(nunique.to_string())

    num_cols = df.select_dtypes(include=[np.number]).columns
    lines.append("\n\nNumeric describe():")
    lines.append(df[num_cols].describe().to_string() if len(num_cols) else "No numeric columns detected.")

    with open(report_path, "w", encoding="utf-8") as f:
        f.write("\n".join(lines))

    print(f"üìù Saved EDA report: {report_path}")


def plot_missing_heatmap(df: pd.DataFrame, max_rows: int = 4000):
    sample = df if len(df) <= max_rows else df.sample(max_rows, random_state=42)
    plt.figure(figsize=(12, 5))
    sns.heatmap(sample.isna(), cbar=False)
    plt.title("Missing Values Heatmap (sampled if large)")
    plt.show()


def plot_outlier_boxplots(df: pd.DataFrame, max_cols: int = 12):
    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    if not num_cols:
        print("No numeric columns for outlier plots.")
        return
    for col in num_cols[:max_cols]:
        plt.figure()
        sns.boxplot(x=df[col])
        plt.title(f"Outlier Check: {col}")
        plt.show()


basic_report(df_raw, REPORT_PATH)
plot_missing_heatmap(df_raw)
plot_outlier_boxplots(df_raw)


# ----------------------------
# 3) PREPROCESSING
# ----------------------------
def normalize_missing_strings(s: pd.Series) -> pd.Series:
    # keep as string, strip, then map common missing tokens to NaN
    s = s.astype(str).str.strip()
    s = s.replace({
        "": np.nan, " ": np.nan,
        "nan": np.nan, "NaN": np.nan,
        "none": np.nan, "None": np.nan,
        "null": np.nan, "NULL": np.nan,
        "na": np.nan, "N/A": np.nan, "n/a": np.nan
    })
    return s


def try_parse_dates(df: pd.DataFrame) -> pd.DataFrame:
    for c in df.columns:
        if df[c].dtype == "object":
            name = str(c).lower()
            if any(k in name for k in ["date", "time", "dt", "timestamp"]):
                parsed = pd.to_datetime(df[c], errors="coerce", infer_datetime_format=True)
                if parsed.notna().mean() >= 0.5:
                    df[c] = parsed
                    print(f"üïí Parsed datetime column: {c}")
    return df


def drop_constant_columns(df: pd.DataFrame) -> pd.DataFrame:
    nunique = df.nunique(dropna=False)
    constant_cols = nunique[nunique <= 1].index.tolist()
    if constant_cols:
        print("üßπ Dropping constant columns:", constant_cols)
        df = df.drop(columns=constant_cols)
    return df


def drop_id_like_columns(df: pd.DataFrame, threshold_ratio: float = 0.98) -> pd.DataFrame:
    n = len(df)
    drop_cols = []
    for c in df.columns:
        unique_ratio = df[c].nunique(dropna=False) / max(n, 1)
        cname = str(c).lower()
        if unique_ratio >= threshold_ratio:
            if any(k in cname for k in ["id", "uid", "key", "code", "number", "no"]):
                drop_cols.append(c)
            else:
                # still drop if it's basically unique and not your target
                if TARGET_COL is None or c != TARGET_COL:
                    drop_cols.append(c)
    if drop_cols:
        print("üßπ Dropping likely ID/high-cardinality columns:", drop_cols)
        df = df.drop(columns=drop_cols)
    return df


def impute_missing(df: pd.DataFrame) -> pd.DataFrame:
    num_cols = df.select_dtypes(include=[np.number]).columns
    cat_cols = df.select_dtypes(include=["object"]).columns
    dt_cols  = df.select_dtypes(include=["datetime64[ns]"]).columns

    # numeric -> median
    for c in num_cols:
        if df[c].isna().any():
            df[c] = df[c].fillna(df[c].median())

    # categorical -> mode
    for c in cat_cols:
        if df[c].isna().any():
            mode = df[c].mode(dropna=True)
            fill_val = mode.iloc[0] if len(mode) else "Unknown"
            df[c] = df[c].fillna(fill_val)

    # datetime: keep NaT (safer than inventing)
    # If you want to fill datetime, uncomment:
    # for c in dt_cols:
    #     if df[c].isna().any():
    #         df[c] = df[c].fillna(df[c].dropna().median())

    return df


def cap_outliers_iqr(df: pd.DataFrame, factor: float = 1.5) -> pd.DataFrame:
    num_cols = df.select_dtypes(include=[np.number]).columns
    for c in num_cols:
        x = df[c].dropna()
        if x.empty:
            continue
        q1, q3 = x.quantile(0.25), x.quantile(0.75)
        iqr = q3 - q1
        if iqr == 0:
            continue
        lo, hi = q1 - factor * iqr, q3 + factor * iqr
        df[c] = df[c].clip(lo, hi)
    return df


def preprocess(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    # Clean object/string cols
    for c in df.select_dtypes(include=["object"]).columns:
        df[c] = normalize_missing_strings(df[c])

    df = try_parse_dates(df)
    df = drop_constant_columns(df)
    df = df.drop_duplicates()
    df = drop_id_like_columns(df, threshold_ratio=0.98)
    df = impute_missing(df)
    df = cap_outliers_iqr(df, factor=1.5)

    return df


df_clean = preprocess(df_raw)
print("\n‚úÖ After preprocessing:")
print("Shape:", df_clean.shape)
display(df_clean.head(10))
df_clean.info()


# ----------------------------
# 4) FEATURE ENGINEERING
# ----------------------------
def feature_engineering(df: pd.DataFrame, target_col: str = None) -> pd.DataFrame:
    out = df.copy()

    # Datetime features
    dt_cols = out.select_dtypes(include=["datetime64[ns]"]).columns.tolist()
    for c in dt_cols:
        out[f"{c}_year"] = out[c].dt.year
        out[f"{c}_month"] = out[c].dt.month
        out[f"{c}_day"] = out[c].dt.day
        out[f"{c}_dayofweek"] = out[c].dt.dayofweek
        out[f"{c}_is_weekend"] = out[c].dt.dayofweek.isin([5, 6]).astype(int)

    # Numeric transforms
    num_cols = out.select_dtypes(include=[np.number]).columns.tolist()
    if target_col in num_cols:
        num_cols.remove(target_col)

    # log1p for non-negative numeric columns
    for c in num_cols:
        if out[c].min() >= 0:
            out[f"log1p_{c}"] = np.log1p(out[c])

    # Frequency encoding for categoricals
    cat_cols = out.select_dtypes(include=["object"]).columns.tolist()
    if target_col in cat_cols:
        cat_cols.remove(target_col)

    for c in cat_cols:
        freq = out[c].value_counts(dropna=False) / len(out)
        out[f"{c}_freq"] = out[c].map(freq)

    # Bin any amount/spend columns
    amount_like = [c for c in out.columns if "amount" in str(c).lower() or "spend" in str(c).lower()]
    for c in amount_like:
        if pd.api.types.is_numeric_dtype(out[c]):
            out[f"{c}_bin"] = pd.qcut(out[c], q=5, duplicates="drop").astype(str)

    return out


df_feat = feature_engineering(df_clean, target_col=TARGET_COL)
print("\n‚úÖ After feature engineering shape:", df_feat.shape)
display(df_feat.head(10))


# ----------------------------
# 5) ONE-HOT ENCODING (ML READY)
# ----------------------------
def make_ml_ready(df: pd.DataFrame, target_col: str = None) -> pd.DataFrame:
    out = df.copy()

    y = None
    if target_col is not None and target_col in out.columns:
        y = out[target_col]
        out = out.drop(columns=[target_col])

    cat_cols = out.select_dtypes(include=["object"]).columns.tolist()
    out_enc = pd.get_dummies(out, columns=cat_cols, drop_first=True)

    if y is not None:
        out_enc[target_col] = y.values

    return out_enc


df_ml = make_ml_ready(df_feat, target_col=TARGET_COL)
print("\n‚úÖ ML-ready shape:", df_ml.shape)
display(df_ml.head(10))


# ----------------------------
# 6) SAVE OUTPUTS
# ----------------------------
df_clean.to_csv(CLEAN_PATH, index=False)
df_ml.to_csv(FEATURES_PATH, index=False)

print("\n‚úÖ Saved outputs to:")
print("Cleaned dataset :", CLEAN_PATH)
print("Features dataset:", FEATURES_PATH)
print("EDA report      :", REPORT_PATH)


# ----------------------------
# 7) TEAM DISCUSSION SUMMARY
# ----------------------------
def discussion_summary(df_before: pd.DataFrame, df_after: pd.DataFrame, df_features: pd.DataFrame):
    print("\n===== TEAM DISCUSSION SUMMARY =====")
    print("Original shape:", df_before.shape)
    print("Cleaned shape :", df_after.shape)
    print("Features shape:", df_features.shape)

    print("\nMissing values (total):")
    print("Before:", int(df_before.isna().sum().sum()))
    print("After :", int(df_after.isna().sum().sum()))

    print("\nTop columns with missing values (before):")
    miss = df_before.isna().sum().sort_values(ascending=False)
    print(miss[miss > 0].head(10) if (miss > 0).any() else "None")

discussion_summary(df_raw, df_clean, df_ml)
```
